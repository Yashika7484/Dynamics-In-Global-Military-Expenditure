## Military Expenditure Trend (2013‚Äì2023)

This visualization shows the **yearly military expenditure** of India and Pakistan from **2013 to 2023**.

## Comparison of Active and Reserve Personnel (Top 5 Militaries ‚Äì 2025)

Military strength is not only determined by technology and expenditure, but also by **manpower** ‚Äî the number of active and reserve personnel available to each nation.  

This visualization compares the **top 5 global militaries** (based on rank) by their **active** and **reserve** personnel counts.  
It provides insight into how countries balance immediate readiness and long-term reserve strength.


```python
# Compare Active vs Reserve Personnel (Top 5)
top5 = df3.nsmallest(5, "rank")
plt.figure(figsize=(15,7))
plt.bar(top5["country"], top5["active_personnel"], label="Active", color="steelblue")
plt.bar(top5["country"], top5["reserve_personnel"], bottom=top5["active_personnel"],
        label="Reserve", color="orange")
plt.title("Active and Reserve Personnel (Top 5 Militaries )", fontsize = 20)
plt.ylabel("Number of Personnel", fontsize = 16)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

### Output
The graph compares the active military personnel among the top five military powers. China leads with the largest number of active troops, followed by India, North Korea, the United States, and Russia. The data shows that while manpower remains a vital defense component, nations like the U.S. and Russia increasingly emphasize technology and modernization over raw personnel strength. India‚Äôs large force size demonstrates its commitment to maintaining readiness and regional security.

### Calculate Average Military Spending (2015‚Äì2019 vs 2020‚Äì2024)
Compute mean defense expenditure for two recent five-year periods to analyze growth trends.


```python
# Calculate average spending from 2015 to 2019
df4["Average_2015_2019"] = (df4["2015"] + df4["2016"] + df4["2017"] + df4["2018"] + df4["2019"]) / 5
```

```python
# Calculate average spending from 2020 to 2024
df4["Average_2020_2024"] = (df4["2020"] + df4["2021"] + df4["2022"] + df4["2023"]+ df4["2024"]) / 5
```

```python
df4.head()
```

### Top 10 Countries by Average Military Spending (2015‚Äì2019)
Identify nations with the highest average defense expenditure during 2015‚Äì2019 to understand global spending priorities in that period.


```python
# Sort by average spending (highest first)
top_10 = df4.sort_values("Average_2015_2019", ascending=False).head(10)

plt.figure(figsize = (15,7))
# Draw a bar chart
plt.plot(top_10["Country"], top_10["Average_2015_2019"], color='indigo', marker = 'o')  
plt.xlabel("Country")
plt.ylabel("Average Spending (2015‚Äì2019)")
plt.title("Top 10 Countries by Average Spending (2015‚Äì2019)", fontsize = 20)
plt.grid(axis = 'y', linestyle = '--')
plt.tight_layout()
plt.show()
```

### Top 10 Countries by Average Military Spending (2020‚Äì2024)
Visualize the countries with the highest average defense budgets during 2020‚Äì2024 to understand recent global military spending trends.


```python
# Sort by average spending (highest first)
top_10 = df4.sort_values("Average_2020_2024", ascending=False).head(10)

plt.figure(figsize = (15,7))
# Draw a bar chart
plt.plot(top_10["Country"], top_10["Average_2020_2024"], color='darkblue', marker = 'o')  
plt.xlabel("Country", fontsize = 16)
plt.ylabel("Average Spending (2020‚Äì2024)", fontsize = 16)
plt.title("Top 10 Countries by Average Spending (2020‚Äì2024)", fontsize = 20)
plt.grid(axis = 'y', linestyle = '--')
plt.tight_layout()
plt.show()
```

### Output
The two charts compare the **average military spending of top countries** during the periods **2015‚Äì2019** and **2020‚Äì2024**.  

In the first period (2015‚Äì2019), **India ranked fifth**, following the United States, China, Russia, and Saudi Arabia. However, in the later period (2020‚Äì2024), India **moved up to fourth position**, overtaking Saudi Arabia in average defense expenditure.  

This upward shift highlights India‚Äôs **strategic focus on military modernization**, **indigenous production**, and **technological advancement** under initiatives like *‚ÄúAtmanirbhar Bharat‚Äù* and *‚ÄúMake in India.‚Äù*  
The post-2020 period also reflects global defense realignments influenced by **border tensions, pandemic recovery, and increasing emphasis on AI, drones, and cyber capabilities**.  

Overall, India‚Äôs rise from 5th to 4th place demonstrates its **growing global defense footprint** and its determination to become a leading power in both regional and international security dynamics.

### Identify Top 10 AI/Drone Technology Seekers
Find the countries or organizations with the highest number of AI drone acquisitions or orders.
